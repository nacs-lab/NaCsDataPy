# Comprehensive Testing

This notebook tests the key features of `SingleData` and ensures things are calculated correctly. 

First, let's use the dataset `20250210_121401`, which is a blast detuning scan of both Na and Cs.

```python
import SingleData
import numpy as np
```

```python
data = SingleData.SingleData('20250210', '121401', 'W:\\NaCsRydLab\\Data')
```

# Plotting Average Images

First, we plot the average images to know what we are dealing with. We will plot images 1 and 2, highlighting sites 1 and 2.

Note: If you want the images to pop out of the notebook use a different backend, such as tk with `%matplotlib tk`

```python
data.plot_avg_img(img_idxs=[1, 2], site_idxs=np.array([1,2]))
```

Note that we can specify different sites and different numbers of sites for each image. This next one plots average images 2 and 3, with sites 1 to 2 and sites 11 to 12 shown in red boxes.

```python
data.plot_avg_img(img_idxs=[2,3], site_idxs=[np.array([1,2]), np.array([11,12])])
```

By default, it plots all images and no sites.

```python
data.plot_avg_img()
```

We can also plot some dual-species images. In the case below, our first image, has combined images 1 and 2, and the second has combined images 3 and 4. Image 1 will be in blue, while image 2 will be in orange.

```python
data.plot_dual_imgs([[1,2], [3,4]], titles = ['Imgs 1 and 2', 'Imgs 3 and 4'])
```

By default, we do not plot any sites, but we can plot sites.

```python
data.plot_dual_imgs([[1,2], [3,4]], site_idxs=[[[1,2], [5,6]], [[], []]],titles = ['Imgs 1 and 2', 'Imgs 3 and 4'])
```

We can also plot the average image as a function of parameter. Notice that this will require loading in all the images. We now look at a dataset that is scanning the AOD frequency (in order to destroy loading in the Cs SLM traps). This dataset will have a different average image per parameter.

```python
data2 = SingleData.SingleData('20250218','111053','W:\\NaCsRydLab\\Data')
```

We use the function `SingleData.plot_avg_img_by_param`, who's full signature is like `plot_avg_img_by_param(self, img_idxs=None, site_idxs=np.array([]), seq_idxs=None, titles=None)`. `img_idxs` is only a single image number that we plot by parameter. `site_idxs` indicates which sites to place boxes around. `seq_idxs` allows for time-binning.

```python
data2.plot_avg_img_by_param(img_idxs=[2])
```
